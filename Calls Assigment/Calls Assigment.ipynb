```python
# importing libs

import pandas as pd
import numpy as np
import matplotlib.pyplot
%matplotlib inline
```

```python
# Read the excel file

df = pd.read_excel("Customer_Call_List.xlsx")
df.head()
```

## Cleaning Data

```python
# rename all column using lower case for ease

df.columns = df.columns.str.lower()
df.head()
```

```python
# add "_" to the column "paying customer" for uniformity

# specify the column
modified_col = "paying customer"

# replace the space in "paying customer" with underscore
df.rename(columns={modified_col: modified_col.replace(" ", "_")}, inplace  = True)

df.head()
```

```python
# Removing all special characters from the phone_number column


df["last_name"] = df["last_name"].str.strip("/_. ") # or str.replace("^[a-zA-Z0-9]", "") (^ means not in)
# repalce null with nothing
df.last_name.head()

```

```python
# Dropin the not usefull column

df = df.drop(columns= "not_useful_column")
df
```

```python
# Turning N/A to blamnks
df.phone_number = df.phone_number.str.replace("N/a", "")

df.head(6)
```

```python
df.phone_number = df.phone_number.fillna("")
df
```

```python
df.phone_number = df.phone_number.str.replace("[^0-9]", "", regex= True)
df.head
```

```python
df
```

```python
# Making values consistent across the do_not_contact column


df.do_not_contact = df.do_not_contact.replace("Y", "Yes").replace("N", "No")
  
df
```

```python
# Making values consistent across the do_not_contact column

df.paying_customer = df.paying_customer.replace("Y", "Yes").replace("N", "No")
df
```

```python
# Droping all rows where "do_not_contact" column is "No"

df = df.drop(df[df["do_not_contact"]== "No"].index)
df
```
